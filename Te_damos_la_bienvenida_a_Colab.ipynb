Minería de Datos:
Es la recolección de información en grandes cantidades. Después se realiza un análisis de información mediante algoritmos de aprendizaje, estrategias específicas. Una vez hecho el análisis de la información, se realiza una interpretación de los datos, detección de patrones, anomalías, etc. Los datos analizados se utilizan ahora para prevenir amenazas, evitar escenarios desfavorables y la evaluación de escenarios futuros.
